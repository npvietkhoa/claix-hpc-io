# Recorder to OTF2

```python
import otf2
import os

def process_trace(otf2_path):
    # Open the OTF2 archive
    with otf2.reader.open(otf2_path) as trace:
        function_times = {}
        for location, event in trace.events:
            if isinstance(event, otf2.events.Enter):
                function = trace.definitions.regions[event.region]
                if function.name not in function_times:
                    function_times[function.name] = 0
            elif isinstance(event, otf2.events.Leave):
                function = trace.definitions.regions[event.region]
                exclusive_time = event.time - event.previous_event.time
                function_times[function.name] += exclusive_time
        return function_times

def process_all_traces(directory):
    all_results = {}
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".otf2"):
                trace_path = os.path.join(root, file)
                print(f"Processing {trace_path}...")
                results = process_trace(trace_path)
                all_results[trace_path] = results
    return all_results

results = process_all_traces("/path/to/otf2/files")
for trace, data in results.items():
    print(f"Trace: {trace}")
    for function, time in data.items():
        print(f"  {function}: {time} ns")
```
